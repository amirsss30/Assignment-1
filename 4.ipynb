```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df = pd.read_excel('price1 (1).xlsx')
plt.scatter(df.area, df.price, color='red', marker='^')
plt.xlabel('area')
plt.ylabel('price')
reg = linear_model.LinearRegression() 
reg.fit(df[['area']],df.price) 
reg.predict([[120]])
reg.predict(df[['area']])
reg.coef_
reg.intercept_
plt.scatter(df.area, df.price, color='red', marker='^')
plt.xlabel('area')
plt.ylabel('price')
plt.plot(df.area, reg.predict(df[['area']]))
pred = pd.read_excel('prediction_price (1).xlsx')
pred.head(3) 
p = reg.predict(pred) 
pred['predicted prices'] = p
pred.to_excel('new.xlsx', index=False) 
plt.show()
```
