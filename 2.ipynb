```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(5493651)
plt.rcdefaults()
fig, ax = plt.subplots()
cities = ('Atyrau', 'Shymkent', 'Astana', 'Almaty', 'Pavlodar')
y_pos = np.arange(len(cities))
performance = 3 + 10 * np.random.rand(len(cities))
error = np.random.rand(len(cities))
ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=cities)
ax.invert_yaxis()
ax.set_xlabel('Temperature')
ax.set_title('The highest temperature at this week:')
plt.show()
```
